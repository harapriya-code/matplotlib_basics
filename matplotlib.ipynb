###  Matplotlib Visualization :

# ========================================
# 1. Introduction to Matplotlib
# ========================================
# Matplotlib is the most widely used Python library for data visualization.
# It allows us to create line plots, scatter plots, bar charts, histograms, pie charts, and more.


```python
# ========================================
# 2. Importing Libraries
# ========================================
import numpy as np                   #  **NumPy** → for numerical data  
import matplotlib.pyplot as plt       # **Matplotlib.pyplot** → for plotting
```

```python
# ========================================
# 3. Basic Line Plot
# ========================================
x = np.linspace(0, 10, 100)   # 100 values between 0 and 10
y = np.sin(x)
#The plot() function creates a simple line plot.  
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# ========================================
# 4. Scatter Plot
# ========================================
x = np.random.rand(50)
y = np.random.rand(50)
#The scatter() function is useful for showing relationships between two variables.  
plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# ========================================
# 5. Bar Chart
# ========================================
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]
# The `bar()` function is useful for categorical data comparisons.
plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# ========================================
# 6. Histogram
# ========================================
data = np.random.randn(1000)
## 📉 Histogram
#The `hist()` function shows the distribution of numerical data.  
#Here we generate 1000 random values and group them into **30 bins**.

plt.hist(data, bins=30, color='green', alpha=0.7)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# ========================================
# 7. Pie Chart
# ========================================
sizes = [20, 30, 25, 25]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
## 🥧 Pie Chart
#The pie() function is useful for showing proportions.  
# autopct displays percentages on the chart.

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

```python
# ========================================
# 8. Subplots
# ========================================
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
## 🪟 Subplots
#- subplot() → manually arrange plots.  
#- subplots() → creates a grid of plots (preferred).  
#Here we plot sine and cosine in separate subplots.

fig, axs = plt.subplots(2, 1, figsize=(6, 6))

axs[0].plot(x, y1, color='blue')
axs[0].set_title("Sine Wave")

axs[1].plot(x, y2, color='orange')
axs[1].set_title("Cosine Wave")

plt.tight_layout()
plt.show()
```

```python
# ========================================
# 9. Displaying Images
# ========================================
img = np.random.rand(10, 10)  # 10x10 random image
## 🖼️ Images
# The `imshow()` function displays image-like data.  
# We also add a colorbar for scale.

plt.imshow(img, cmap='viridis')
plt.title("Image Example")
plt.colorbar()
plt.show()
```

```python
# ========================================
# 10. Dataset Visualization
# ========================================

# Simulated dataset
months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun"])
sales = np.array([150, 200, 250, 220, 300, 400])
profit = np.array([50, 70, 90, 85, 120, 160])

# Line Plot (Sales over months)
plt.plot(months, sales, marker='o', label="Sales")
plt.plot(months, profit, marker='s', label="Profit")
plt.title("Sales & Profit Over Time")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.show()

# Bar Plot (Sales comparison)
plt.bar(months, sales, color='skyblue')
plt.title("Monthly Sales (Bar Chart)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# Scatter Plot (Sales vs Profit)
plt.scatter(sales, profit, color='red')
plt.title("Sales vs Profit (Scatter Plot)")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
